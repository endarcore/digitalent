##Analisa data kemiskinan Yogyakarta, dari tahun 2007 hingga tahun 2013

```python
import pandas as pd
```

```python
data_kemiskinan = pd.read_csv('data_kemiskinan_jogja.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_kemiskinan
```

```python
data_kemiskinan.columns
```

```python
data_kemiskinan[['nama_kabkota','kecamatan','tahun','kategori']]
```

```python
data_kemiskinan.kategori.count()
```

```python
data_kemiskinan.groupby('kategori').mean()
```

```python
ax = data_kemiskinan.groupby('kategori').mean()
ax = ax.plot(kind='bar', title ="Masyarakat Miskin Yogyakarta",figsize=(10,10), legend=True, fontsize=10)
ax.set_xlabel("kategori", fontsize=10)
ax.set_ylabel("Jumlah", fontsize=10)
```

##Analisis

Berdasarkan data tersebut dapat ditarik kesimpulan bahwa mayoritas penduduk Yogyakarta berada dalam kategori tidak miskin, dilihat dari total penduduk dengan perbandingan jumlah kemiskinan, tahun 2007 sampai 2013
